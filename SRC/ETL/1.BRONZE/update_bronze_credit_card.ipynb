```python
df = spark.read.csv("/Volumes/big_data/raw/data/creditcard.csv", header=True, inferSchema=True)
display(df)
```

```python
# Análise de distribuição da coluna de fraude (Class)
import pyspark.sql.functions as F

resumo_classes = df.groupBy('Class').count().orderBy('Class')
display(resumo_classes)
```

```python
# Estatísticas descritivas das variáveis, separando fraudes e legítimas
import pandas as pd
columns = df.columns
colunas_reais = [col for col in columns if col not in ['Time', 'Class']]

stats_fraude = df.filter(df.Class == 1).select(colunas_reais).describe().toPandas()
stats_fraude['fraude'] = 'Sim'
stats_legit = df.filter(df.Class == 0).select(colunas_reais).describe().toPandas()
stats_legit['fraude'] = 'Não'
resumo = pd.concat([stats_fraude, stats_legit])
display(resumo)
```

```python
# Análise descritiva separada por tipo de transação
columns = df.columns
colunas_reais = [col for col in columns if col not in ['Time', 'Class']]

# Estatísticas para fraudes
stats_fraude = df.filter(df.Class == 1).select(colunas_reais).describe().toPandas()
stats_fraude['fraude'] = 'Sim'

# Estatísticas para não fraudes
stats_legit = df.filter(df.Class == 0).select(colunas_reais).describe().toPandas()
stats_legit['fraude'] = 'Não'

# Junta resultados para facilitar comparação
resumo = pd.concat([stats_fraude, stats_legit])
display(resumo)
```
